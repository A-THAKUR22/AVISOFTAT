Student database

```python
class Student:
    def __init__(self):
        self.students = []

    def add_student_description(self, name, roll_no, marks):
        student = {'Name': name, 'Roll No.': roll_no, 'Marks': marks}
        self.students.append(student)

    def display_student_details(self):
        for student in self.students:
            print(f"Name: {student['Name']}")
            print(f"Roll Number: {student['Roll No.']}")
            print(f"Marks: {student['Marks']}\n")

    def max_min_thres(self):
        max_threshold = 90
        min_threshold = 40

        for student in self.students:
            student_name = student['Name']
            for subject, marks in student['Marks'].items():
                if marks >= max_threshold:
                    print(f"{student_name} performed great in {subject} since {marks} have surpassed the threshold.")
                elif marks >= min_threshold:
                    print(f"{student_name} performed good in {subject} since {marks}, can do better.")
                else:
                    print(f"{student_name} didn't perform well in {subject} and needs personal attention.")
        sorted_students = sorted(self.students, key=lambda x: sum(x['Marks'].values()), reverse=True)
        print("\nTop 5 Toppers based on total marks:")
        for i, student in enumerate(sorted_students[:5]):
            total_marks = sum(student['Marks'].values())
            print(f"{i + 1}. {student['Name']} - Total Marks: {total_marks}")
            if i == 0:
                print(f"{student['Name']} is the top scorer and will be rewarded separately.")

    def top_ranker(self):
        sorted_students = sorted(self.students, key=lambda x: sum(x['Marks'].values()), reverse=True)
        topper = sorted_students[0]['Name']
        print(f"{topper} is the first rank holder and will be awarded with a certificate of appreciation and a trophy.")
    def calculate_average(self):
      above_threshold = []
      below_threshold = []
      mid = []
      total_marks = {subject: sum(student['Marks'][subject] for student in self.students) for subject in self.students[0]['Marks']}
      average_marks = {subject: total / len(self.students) for subject, total in total_marks.items()}
      for student in self.students:
        for subject, marks in student['Marks'].items():
            if marks > average_marks[subject] + 10:
                above_threshold.append(f"{student['Name']} in {subject}")
            elif marks >= average_marks[subject] - 10:
                mid.append(f"{student['Name']} in {subject}")
            else:
                below_threshold.append(f"{student['Name']} in {subject}")
      print(f"(i) Above max threshold: {above_threshold}")
      print(f"(ii)  Mid threshold: {mid}")
      print(f"(iii) Below min threshold: {below_threshold}")


class WeakStudent(Student):
    def personal_attent(self):
        super().calculate_average()
        min_threshold = 40
        personal_attention = []

        for student in self.students:
            if student['Marks'] < min_threshold:
                personal_attention.append(student['Name'])

        print(f"These students need personal attention: {personal_attention}")


class AnalysisOptions:
    def __init__(self, options, data):
        self.used_class = data

        if options == 1:
            self.used_class.display_student_details()
        elif options == 2:
            self.used_class.max_min_thres()
        elif options == 3:
            self.used_class.top_ranker()
        elif options == 4:
            weak_student_instance = WeakStudent()
            weak_student_instance.students = self.used_class.students
            weak_student_instance.personal_attent()
        else:
            print("Make a valid choice!!!")






database = Student()
while True:
  try:
    name = input("Enter the student name:")
    if name.lower() == "exit":
        break
    roll_number = int(input("Enter the roll number:"))

    marks_input = input("Enter subject marks separated by spaces (e.g., Maths:99 English:77):")
    marks_dict = dict(items.split(":") for items in marks_input.split())
    marks = {subject: int(mark) for subject, mark in marks_dict.items()}

    database.add_student_description(name, roll_number, marks)
  except ValueError as e:
    print(f"Error: {e}. Please enter valid input.")
while True:
  try:
    option_value = int(input("Enter the option corresponding to the data you need:"
                         "\n1. Display student details"
                         "\n2. Students above maximum and below minimum threshold"
                         "\n3. Top Rank holder of the class"
                         "\n4. Students in need of personal attention"
                         "\n5. Exit\n"))
  except ValueError :
    print("Error: Please input a valid integer option.")
  if option_value == 5:
    exit()
    try:
      my_choice = AnalysisOptions(option_value, database)
    except IndexError:
      print("Error: Invalid option. Please make a valid choice!!!")

```

Plot graphs for company sales data of the company

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv('company_sales_data.csv')
data
```

1. Read Total profit of all months and show it using a
line plot

```python
x=data['month_number']
y=data['total_profit']
plt.yticks(np.array([100000,200000,300000,400000,500000]))
plt.plot(x,y,'-')
plt.title('Company profit per month')
plt.xlabel('Month number')
plt.ylabel('Profit in dollar')
```

2. Get total profit of all months and show line plot with
   the following Style properties

```python
x=data['month_number']
y=data['total_units']
plt.plot(x,y,'r--o',mfc='b',label='Profit data of last year')

plt.title('Company sale data last year')
plt.xlabel('Month Number')
plt.ylabel('Profit in dollar')
plt.legend(loc='lower right')

```

3. Read all product sales data and show it using a
   multiline plot

```python
x=data['month_number']
plt.xticks(x)
plt.yticks(np.array([1000,2000,4000,6000,8000,10000,12000,15000,18000]))
plt.plot(x,data['facecream'],'b-o',label='Face cream Sales Data')
plt.plot(x,data['facewash'],'g-o',label='Face wash Sales Data')
plt.plot(x,data['toothpaste'],'r-o',label='Toothpaste Sales Data')
plt.plot(x,data['bathingsoap'],'y-o',label='Bathing soap Sales Data')
plt.plot(x,data['shampoo'],'o-',color='brown',label='Shampoo Sales Data')
plt.title('Sales data')
plt.xlabel('Month Number')
plt.ylabel('Sales units in numbers')
plt.legend()
```

4. Read toothpaste sales data of each month and show
it using a scatter plot

```python
x=data['month_number']
y=data['toothpaste']
plt.xticks(x)
plt.yticks(np.array([4500,5000,5500,6000,6500,7000,7500,8000]))
plt.scatter(x,y,label='Tooth paste Sales data')
plt.grid(linestyle='dotted')
plt.title("Tooth paste sales data")
plt.xlabel('Month Number')
plt.ylabel('Number of units Sold')
plt.legend()
```

5. Read face cream and facewash product sales data
and show it using the bar chart

```python
x=data['month_number']
x1=np.arange(len(x))
x2=[x2 +0.35 for x2 in x1]
plt.xticks([x + 0.35/ 2 for x in range(len(x))], x)
plt.yticks(np.array([500,1000,1500,2000,2500,3000,3500]))
plt.bar(x,data['facecream'],color='blue',width=0.35,label='Face Cream sales data')
plt.bar(x2,data['facewash'],color='orange',width=0.35,label='Face Wash sales data')
plt.grid(linestyle='dotted')
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.legend(loc='upper left')
```

6. Read sales data of bathing soap of all months and
   show it using a bar chart. Save this plot to your hard
  disk

```python
x=data['month_number']
plt.xticks(x)
plt.bar(x,data['bathingsoap'],width=0.7)
plt.grid(linestyle='dotted')
plt.xlabel('Month Number')
plt.ylabel('Sales units in numbers')
plt.title('bathingsoap sales data')
```

7. Read the total profit of each month and show it using
   the histogram to see the most common profit ranges

```python
x=data['total_profit']
plt.xticks(np.array([150000,175000,200000,225000,250000,300000,350000]))
plt.yticks(np.array([0,1,2,3,4,5]))
plt.hist(x,label='Profit data')
plt.xlabel('profit range in dollar')
plt.ylabel('Actual profit in dollar')
plt.legend()
```

Calculate total sale data for last year for each
product and show it using a Pie chart

```python
sale=[data['facecream'].sum(),data['facewash'].sum(),data['toothpaste'].sum(),data['bathingsoap'].sum(),data['shampoo'].sum(),data['moisturizer'].sum()]
s_labels=['Facecream','Facewash','ToothPaste','BathingSoap','Shampoo','Moisturizer']
plt.pie(sale,labels=s_labels,autopct="%0.1f%%")
plt.title("Sales Data")
plt.legend(loc="lower right")
```

Read Bathing soap facewash of all months and
display it using the Subplot

```python
month=data['month_number'].tolist()
b_soap=data['bathingsoap'].tolist()
f_wash=data['facewash'].tolist()
f,axarr = plt.subplots(2,sharex=True)
axarr[0].plot(month,b_soap,label='Bathing Soap Sales Date',color='k',marker='o',linewidth=3)
axarr[0].set_title('Sales data of a Bathing Soap')
axarr[1].plot(month,f_wash,label='Face Wash Sales Date',color='r',marker='o',linewidth=3)
axarr[1].set_title('Sales data of a facewash')
plt.xticks(month)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
```

Read all product sales data and show it using the
stack plot

```python
x=data['month_number']
plt.xticks(x)
plt.stackplot(x,data['facecream'],data['facewash'],data['toothpaste'],data['bathingsoap'],data['shampoo'],data['moisturizer'],labels=['Facecream','Facewash','Toothpaste','BathingSoap','Shampoo','Moisturizer'])
plt.title('Sales data')
plt.xlabel('Month Number')
plt.ylabel('Sales units in numbers')
plt.legend(loc='upper left')
```

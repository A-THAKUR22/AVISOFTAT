Plot a graph using matplotlib based on the given points

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x=[0,1,2,3,4]
y=[0,2,4,6,8]
x2=np.arange(0,4.5,0.5)
plt.plot(x,y,'b^--',label="2x")
plt.plot(x2[:6],x2[:6]**2,'r-',label="x^2")
plt.plot(x2[5:],x2[5:]**2,'r--')
plt.title("Our First Graph!")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
```

Plot the graph based on the values of the variables in the .csv file

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data=pd.read_csv('gas_prices.csv')
x=data['Year']
plt.plot(x,data['Australia'],'y-o')
plt.plot(x,data['Canada'],'r-o')
plt.plot(x,data['USA'],'b-o',label='United States')
plt.plot(x,data['Japan'],'g-o')

plt.xlabel("Year")
plt.ylabel("Countries gas prices")
plt.legend()

```
